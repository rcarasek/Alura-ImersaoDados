```python
import pandas as pd
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
```

```python
 dados['tratamento'].unique()
```

```python
mapa = {'droga': 'composto'}   #SmallMolecules
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts()[:5]   #do 0 ao 4
```

```python
cod_compostos
```

```python
cod_compostos.index
```

```python
dados_5mais = dados.query('composto in @cod_compostos.index')
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
sns.countplot(x = 'composto', data = dados_5mais)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(12,3))
sns.countplot(x = 'composto', data = dados_5mais)
```
